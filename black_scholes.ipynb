# Black Scholes Model

More Info: [Wiki](https://en.wikipedia.org/wiki/Black%E2%80%93Scholes_model)

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def integrate(f, a, b, steps = 1000):
    area = 0

    step_size = (b - a) / steps

    for i in range(steps):
        area += f(a + i * step_size)

    return (area * step_size)

def normalCumulativeDistributionFunction(x : float):
    # Approximation of the normal cumulative distribution function
    start = -5
    end = x
    steps = 10_000

    area = 0

    step_size = (end - start) / steps

    for i in range(steps):
        area += np.exp(-0.5 * (start + i * step_size) ** 2)

    area *= step_size
    
    return (area / np.sqrt(2 * math.pi))

def probabilityDensityFunction(x : float):
    return (1 / np.sqrt(2 * math.pi)) * np.exp(-0.5 * x ** 2)
```

```python
plot_x = np.arange(-4, 4, 0.1)
plot_y = [probabilityDensityFunction(x) for x in plot_x]

plt.scatter(plot_x, plot_y)
plt.show()

plot_x = np.arange(-4, 4, 0.1)
plot_y = [normalCumulativeDistributionFunction(x) for x in plot_x]

plt.scatter(plot_x, plot_y)
plt.show()
```
